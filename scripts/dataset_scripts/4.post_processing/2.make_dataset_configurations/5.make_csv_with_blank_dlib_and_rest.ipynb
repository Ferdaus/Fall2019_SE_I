```python
class_names = ['mirror_selfie','non_mirror_selfie','non_selfie','selfie']
```

```python
class_name = class_names[3]
```

```python
ims_ypr_dlib = pd.read_csv(ims_ypr_dlib_csv_base_path+class_name+'.csv')
```

```python
ims_ypr_dlib.head()
```

```python
ims_ypr_dlib['dlib'] = ims_ypr_dlib.im_name.replace('.jpg','_dlib_only.jpg', inplace = False, regex = True)
```

```python
ims_ypr_dlib['im_dlib'] = ims_ypr_dlib.im_name.replace('.jpg','_dlib.jpg', inplace = False, regex = True)
```

```python
ims_ypr_dlib['skel_dlib'] = ims_ypr_dlib.skeleton.replace('_skeleton.jpg','_skeleton_dlib.jpg', inplace = False, regex = True)
```

```python
ims_ypr_dlib['im_skel_dlib'] = ims_ypr_dlib.im_skeleton.replace('_im_skeleton.jpg','_im_skeleton_dlib.jpg', inplace = False, regex = True)
```

```python
ims_ypr_dlib.head()
```

```python
ims_ypr_dlib.to_csv(save_csv_base_path+f'{class_name}.csv',
                              columns = ims_ypr_dlib.columns, 
                              header = ims_ypr_dlib.columns,
                              index = False)
```
