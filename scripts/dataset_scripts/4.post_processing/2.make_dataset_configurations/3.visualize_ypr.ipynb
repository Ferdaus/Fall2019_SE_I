```python
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import os
import scipy.io as sio
import cv2
import math
from math import cos, sin
import time
```

```python
csv_path = '/disks/data/paper_projects/selfie_detection/data/DS_224_all_features_no_duplicates/yaw_pitch_roll/'


class_name = 'mirror_selfie'

image_path = '/disks/data/paper_projects/selfie_detection/data/DS_224_all_features_no_duplicates/raw_image/'+class_name+'/'

filename = f'{class_name}.csv'
im_info_df = pd.read_csv(csv_path+filename)
```

```python
len(im_info_df.index), im_info_df.head()
```

```python
def draw_axis(img, yaw, pitch, roll, tdx=None, tdy=None, size = 100):

    pitch = pitch * np.pi / 180
    yaw = -(yaw * np.pi / 180)
    roll = roll * np.pi / 180

    if tdx != None and tdy != None:
        tdx = tdx
        tdy = tdy
    else:
        height, width = img.shape[:2]
        tdx = width / 2
        tdy = height / 2

    # X-Axis pointing to right. drawn in red
    x1 = size * (cos(yaw) * cos(roll)) + tdx
    y1 = size * (cos(pitch) * sin(roll) + cos(roll) * sin(pitch) * sin(yaw)) + tdy

    # Y-Axis | drawn in green
    #        v
    x2 = size * (-cos(yaw) * sin(roll)) + tdx
    y2 = size * (cos(pitch) * cos(roll) - sin(pitch) * sin(yaw) * sin(roll)) + tdy

    # Z-Axis (out of the screen) drawn in blue
    x3 = size * (sin(yaw)) + tdx
    y3 = size * (-cos(yaw) * sin(pitch)) + tdy

    cv2.line(img, (int(tdx), int(tdy)), (int(x1),int(y1)),(0,0,255),3)
    cv2.line(img, (int(tdx), int(tdy)), (int(x2),int(y2)),(0,255,0),3)
    cv2.line(img, (int(tdx), int(tdy)), (int(x3),int(y3)),(255,0,0),2)

    return img
```

```python
sample_count = 10

idxs = np.random.randint(0,len(im_info_df.index),(sample_count))

for idx in idxs:
    row = im_info_df.iloc[idx]
    fname = row.filename
    yaw = row.yaw
    pitch = row.pitch
    roll = row.roll
    x_min = row.x_min
    x_max = row.x_max
    y_min = row.y_min
    y_max = row.y_max
    bbox_height = abs(y_max - y_min)
    im = cv2.imread(image_path+fname)
    im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
    
    drawn_im = draw_axis(im, yaw, pitch, roll, tdx = (x_min + x_max) / 2, tdy= (y_min + y_max) / 2, size = bbox_height/2)
    
    plt.imshow(drawn_im)
    plt.show()
    time.sleep(2)
```
